```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles.csv")
print(df.shape)
print(df.head())
```

```python
print(df.isnull().sum())
df.dropna(inplace=True) 
print(df.shape)
```

```python
df['type'] = df['type'].dropna()
type_counts = df['type'].value_counts()
type_counts.plot(kind='bar' , color=['red' , 'purple'][:len(type_counts)])
plt.title("Movies vs tv shows")
plt.xlabel("Type")
plt.ylabel("count")
plt.grid(axis='y')
plt.show()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line', color='green')
plt.title("content over time")
plt.xlabel("year")
plt.ylabel("number of items added")
plt.grid(True)
plt.show()
```

```python
df['country'].value_counts().head(10).plot(kind='bar',color='orange')
plt.title("Top 10 countries with most contents")
plt.xlabel("country")
plt.ylabel("number of titles")
plt.show()
```

```python
df['rating'].value_counts().plot(kind='pie' , autopct='%1.1f%%')
plt.title("Rating distribution")
plt.show()
```
